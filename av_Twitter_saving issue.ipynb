```python
ind=1
import requests
import json
import time
import pandas as pd
from datetime import datetime

# No BEARER_TOKEN is given for testing
BEARER_TOKEN = ""

output_ls = []

def get_tweets(screen_name, user_id, npi_id, standard_author, user_index,ind, until_id):
    num_ids_lst = []
    ind=ind+1
    print("user_id!!", user_id,ind)

    max_tweets = 100
    start_time = "2024-10-01"
    end_time = "2024-12-31"

    if until_id:
        url = f"https://api.twitter.com/2/users/{user_id}/tweets?max_results={max_tweets}&start_time={start_time}T00:00:00Z&end_time={end_time}T00:00:00Z&until_id={until_id}"
    else:
        url = f"https://api.twitter.com/2/users/{user_id}/tweets?max_results={max_tweets}&start_time={start_time}T00:00:00Z&end_time={end_time}T00:00:00Z"

    headers = {
        'Authorization': f'Bearer {BEARER_TOKEN}',  # No token for testing
    }

    try:
        response = requests.get(url, headers=headers)
        resp_ = json.loads(response.text)

        if response.status_code == 401:
            raise Exception("Unauthorized access. Missing or invalid BEARER_TOKEN.")

        if "errors" in resp_:
            raise Exception(resp_["errors"][0]["message"])

        if resp_["meta"]["result_count"] == 0:
            return

    except Exception as e:
        print(f"Error occurred: {e}")
        error_message = "Error has found, please try again later"

        # **Filling all columns with error message**
        data_dict = {col: error_message for col in master_columns}
        output_ls.append(data_dict)

        # **Saving to temp file**
        data_df = pd.DataFrame(output_ls)
          # Unique timestamp for file name
        data_df.to_excel(f"/Users/vineethaanumula/documents/cursor/output/Twitter_part1_temp_outputs{timestamp}.xlsx")

        print("Completed***********************")  # Marking process as done
        return

    print("Fetching completed successfully!")  # This prints only if no error occurs

def chunked_processing():
    chunk_size = 100
    for i in range(0, len(handle_lst), chunk_size):
        chunk = handle_lst[i:i+chunk_size]
        for user_handle in chunk:
            user_index = handle_lst.index(user_handle)
            if status_lst[user_index] == 'notcompleted':
                get_tweets(user_handle, user_id_lst[user_index], npi_id_lst[user_index], standard_author_lst[user_index],ind, user_index, "")

chunked_processing()
```
